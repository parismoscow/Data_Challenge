```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These/path)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load, encoding="utf8")
```

```python
Total_players = len(purchase_data_df["SN"].value_counts())
purchasing_summary_df = pd.DataFrame({"Total Players":[Total_players]})
purchasing_summary_df
```

```python
Number_of_Unique_Items = len(purchase_data_df["Item ID"].value_counts())
```

```python
purchase_data_df.describe()
```

```python
Total_Revenue = purchase_data_df["Price"].sum()
```

```python
Number_of_Purchases = len(purchase_data_df["Purchase ID"].value_counts())
```

```python
Average_Price = purchase_data_df["Price"].mean()
```

```python
purchasing_summary_df = pd.DataFrame({"Number of Unique Items":[Number_of_Unique_Items],
                                             "Average Purchase Price":[Average_Price],
                                            "Total Number of Purchases":[Number_of_Purchases],
                                             "Total Revenue":[Total_Revenue]})
purchasing_summary_df
purchasing_summary_df["Average Purchase Price"]=purchasing_summary_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)
purchasing_summary_df["Total Revenue"]=purchasing_summary_df["Total Revenue"].astype(float).map("${:,.2f}".format)
purchasing_summary_df
```

```python
Gender_Demographic = purchase_data_df.groupby("Gender")["SN"].nunique()
```

```python
#Percentage of Players
Total_players = len(purchase_data_df["SN"].value_counts())
Percentage_of_Players= (Gender_Demographic/Total_players)*100
```

```python
purchasing_summary_df = pd.DataFrame({"Total Count":Gender_Demographic,
                                      "Percentage of Players":Percentage_of_Players})
purchasing_summary_df
```

```python
#Purchase Count
purchase_count_by_gender = purchase_data_df.groupby("Gender")["Purchase ID"].nunique()
```

```python
#Average Purchase Price
Average_Purchase_Price_by_gender = purchase_data_df.groupby("Gender")["Price"].mean()
```

```python
#Total Purchase Value
Total_Purchase_Value_by_gender = purchase_data_df.groupby("Gender")["Price"].sum()
```

```python
#Avg Total Purchase per Person
Total_Purchase_Value_by_gender = purchase_data_df.groupby("Gender")["Price"].sum()
Gender_Demographic = purchase_data_df.groupby("Gender")["SN"].nunique()
Avg_Total_Purchase_per_Person_by_gender = Total_Purchase_Value_by_gender/Gender_Demographic
```

```python
purchasing_summary_df = pd.DataFrame({"Purchase Count":purchase_count_by_gender,
                                      "Average Purchase Price":Average_Purchase_Price_by_gender,
                                     "Total Purchase Value":Total_Purchase_Value_by_gender,
                                     "Avg Total Purchase per Person":Avg_Total_Purchase_per_Person_by_gender})
purchasing_summary_df["Average Purchase Price"]=purchasing_summary_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)
purchasing_summary_df["Total Purchase Value"]=purchasing_summary_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
purchasing_summary_df["Avg Total Purchase per Person"]=purchasing_summary_df["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)
purchasing_summary_df
```

```python
#Age Demographics
#Establish bins for ages <10, 10-14, 15-19, 20-24, 25-29, 30-34, 35-39, 40+
bins =[0, 9, 14, 19, 24, 29, 34, 39, 999]
bins_for_ages = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", ">=40"]
#Categorize the existing players using the age bins
purchase_data_df["Categorized players"]= pd.cut(purchase_data_df["Age"],bins, labels=bins_for_ages)
#Calculate the numbers and percentages by age group
age_grouped = purchase_data_df.groupby("Categorized players")
count_age = age_grouped["SN"].nunique()
percentage_by_age = (count_age/Total_players) * 100
age_demographics = pd.DataFrame({"Percentage of Players": percentage_by_age, "Total Count": count_age})
age_demographics.index.name = None
age_demographics
```

```python
#Purchasing Analysis (Age)
#Bin the purchase_data data frame by age
# Run basic calculations to obtain purchase count, avg. purchase price, 
#avg. purchase total per person
purchase_count_age = age_grouped["Purchase ID"].count()

average_purchase_price_age = age_grouped["Price"].mean()

total_purchase_value = age_grouped["Price"].sum()

average_purchase_per_person_age = total_purchase_value/count_age

# Create a summary data frame to hold the results
age_demographics = pd.DataFrame({"Purchase Count": purchase_count_age,
                                 "Average Purchase Price": average_purchase_price_age,
                                 "Total Purchase Value":total_purchase_value,
                                 "Average Purchase Total per Person": average_purchase_per_person_age})

# Display the summary data frame
age_demographics.index.name = None
age_demographics["Average Purchase Price"]= age_demographics["Average Purchase Price"].astype(float).map("${:,.2f}".format)
age_demographics["Total Purchase Value"]= age_demographics["Total Purchase Value"].astype(float).map("${:,.2f}".format)
age_demographics["Average Purchase Total per Person"]= age_demographics["Average Purchase Total per Person"].astype(float).map("${:,.2f}".format)
age_demographics
```

```python
#Top Spenders
#Calculate Purchase Count	Average Purchase Price	Total Purchase Value
buyers = purchase_data_df.groupby("SN")
count_buyers = buyers["Purchase ID"].count()
avg_purchase_price_buyers = buyers["Price"].mean()
purchase_total_buyers = buyers["Price"].sum()

#Create a summary data frame to hold the results
#Sort the total purchase value column in descending order
top_spenders = pd.DataFrame({"Purchase Count": count_buyers,
                             "Average Purchase Price": avg_purchase_price_buyers,
                             "Total Purchase Value":purchase_total_buyers})
filtered_buyers = top_spenders.sort_values(["Total Purchase Value"], ascending=False).head()
filtered_buyers["Average Purchase Price"]=filtered_buyers["Average Purchase Price"].astype(float).map("${:,.2f}".format)
filtered_buyers["Total Purchase Value"]=filtered_buyers["Total Purchase Value"].astype(float).map("${:,.2f}".format)

filtered_buyers
```

```python
#Most Popular Items
items = purchase_data_df[["Item ID", "Item Name", "Price"]]
items_grouped = purchase_data_df.groupby(["Item ID","Item Name"])
purchase_count = items_grouped["Price"].count()
purchase_price = items_grouped["Price"].sum()
per_item = purchase_price/purchase_count
popular_items = pd.DataFrame({"Purchase Count": purchase_count, 
                                   "Item Price": per_item,
                                   "Total Purchase Value":purchase_price})
popular_items_sorted = popular_items.sort_values(["Purchase Count"], ascending=False)
popular_items_sorted["Item Price"]=popular_items_sorted["Item Price"].astype(float).map("${:,.2f}".format)
popular_items_sorted["Total Purchase Value"]=popular_items_sorted["Total Purchase Value"].astype(float).map("${:,.2f}".format)

popular_items_sorted.head()
```

```python
#Most Profitable Items

#Sort the above table by total purchase value in descending order
popular_items_sorted = popular_items.sort_values(["Total Purchase Value"], ascending=False)
popular_items_sorted["Item Price"]=popular_items_sorted["Item Price"].astype(float).map("${:,.2f}".format)
popular_items_sorted["Total Purchase Value"]=popular_items_sorted["Total Purchase Value"].astype(float).map("${:,.2f}".format)
popular_items_sorted.head()
```
